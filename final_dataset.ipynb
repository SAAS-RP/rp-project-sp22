This file is used to stratify works and words by year, so that the dataset can be deconstructed

```python
import pandas as pd
import numpy as np
import collections
import pickle
from IPython.display import display
```

Deserializing DataFrame from basic_dataset.py

```python
words_with_years = pd.read_pickle('./words.pkl')
years_and_books = pd.read_pickle('./books.pkl')
sentences_with_years = pd.read_pickle('./sentences.pkl')
display(words_with_years)
display(years_and_books)
display(sentences_with_years)
```

Sorting each row by year in order to create proper time period bins.

```python
words_with_years = words_with_years.sort_index()
display(words_with_years)
years_and_books = years_and_books.sort_index()
display(years_and_books)
sentences_with_years = sentences_with_years.sort_index()
display(sentences_with_years)
```

Finding evenly ditributed time period bins, by number of works in those years, and stratifying accordingly.

```python
year_freqs = []
total_sum = 0
for row in years_and_books.iterrows():
    freq = 0
    for elem in row[1].iteritems():
        if elem[1] is None:
            year_freqs.append([row[0], freq])
            break
        else:
            freq += 1
            total_sum += 1
current_bin = 1
num_bins = 4
bins = []
ind = 0
start = 0
end = 0
temp_sum = 0
while ind < len(year_freqs)-1:
    temp_sum += year_freqs[ind][1]
    if temp_sum > current_bin*total_sum/num_bins:
        end = ind
        bins.append([year_freqs[start][0], year_freqs[end][0]])
        start = ind+1
        current_bin += 1
    ind += 1
bins.append([year_freqs[start][0], year_freqs[ind][0]])
print(bins)






        
```

These 4 bins most closely correspond to existing eras of literature, those being:

Classical/Neoclassical: before 1660, 1660-1790
Romantic: 1790-1830
Victorian: 1932-1901
Modern/Postmodern: 1914-Present

Source: https://www.summitlearning.org/docs/227310

Now that we have our bins, we can separate the existing dataset into these 4 categories.

```python
classical_b = years_and_books.loc[bins[0][0]:bins[0][1]]
classical_w = words_with_years.loc[bins[0][0]:bins[0][1]]
classical_s = sentences_with_years.loc[bins[0][0]:bins[0][1]]

romantic_b = years_and_books.loc[bins[1][0]:bins[1][1]]
romantic_w = words_with_years.loc[bins[1][0]:bins[1][1]]
romantic_s = sentences_with_years.loc[bins[1][0]:bins[1][1]]

victorian_b = years_and_books.loc[bins[2][0]:bins[2][1]]
victorian_w = words_with_years.loc[bins[2][0]:bins[2][1]]
victorian_s = sentences_with_years.loc[bins[2][0]:bins[2][1]]

modern_b = years_and_books.loc[bins[3][0]:bins[3][1]]
modern_w = words_with_years.loc[bins[3][0]:bins[3][1]]
modern_s = sentences_with_years.loc[bins[3][0]:bins[3][1]]
```

Then we separate these into NumPy arrays to consolidate them into a single dataset of books and words according to the era.

```python
classical_bs = classical_b.to_numpy()
classical_ws = classical_w.to_numpy()
classical_ss = classical_s.to_numpy()

romantic_bs = romantic_b.to_numpy()
romantic_ws = romantic_w.to_numpy()
romantic_ss = romantic_s.to_numpy()

victorian_bs = victorian_b.to_numpy()
victorian_ws = victorian_w.to_numpy()
victorian_ss = victorian_s.to_numpy()

modern_bs = modern_b.to_numpy()
modern_ws = modern_w.to_numpy()
modern_ss = modern_s.to_numpy()

#final classical dataset
classical = collections.defaultdict()
class_b = np.array([])
for b in classical_bs:
    b = [x for x in b if x is not None]
    class_b = np.concatenate((class_b, b))
class_w = np.array([])
for w in classical_ws:
    w = [x for x in w if x is not None]
    class_w = np.concatenate((class_w, w))
class_w = np.unique(class_w)
class_s = np.array([])
for s in classical_ss:
    s = [x for x in w if s is not None]
    class_s = np.concatenate((class_s, s))
    
classical['words'] = class_w
classical['books'] = class_b
classical['sentences'] = class_s

#final romantic dataset
romantic = collections.defaultdict()
rom_b = np.array([])
for b in romantic_bs:
    b = [x for x in b if x is not None]
    rom_b = np.concatenate((rom_b, b))
rom_w = np.array([])
for w in romantic_ws:
    w = [x for x in w if x is not None]
    rom_w = np.concatenate((rom_w, w))
rom_w = np.unique(rom_w)
rom_s = np.array([])
for s in romantic_ss:
    s = [x for x in s if x is not None]
    rom_s = np.concatenate((rom_s, s))
    
romantic['words'] = rom_w
romantic['books'] = rom_b
romantic['sentences'] = rom_s

#final victorian dataset
victorian = collections.defaultdict()
vic_b = np.array([])
for b in victorian_bs:
    b = [x for x in b if x is not None]
    vic_b = np.concatenate((vic_b, b))
vic_w = np.array([])
for w in victorian_ws:
    w = [x for x in w if x is not None]
    vic_w = np.concatenate((vic_w, w))
vic_w = np.unique(vic_w)
vic_s = np.array([])
for s in victorian_ss:
    s = [x for x in s if x is not None]
    vic_s = np.concatenate((vic_s, s))
    
victorian['words'] = vic_w
victorian['books'] = vic_b
victorian['sentences'] = vic_s

#final modern dataset
modern = collections.defaultdict()
mod_b = np.array([])
for b in modern_bs:
    b = [x for x in b if x is not None]
    mod_b = np.concatenate((mod_b, b))
mod_w = np.array([])
for w in modern_ws:
    w = [x for x in w if x is not None]
    mod_w = np.concatenate((mod_w, w))
mod_w = np.unique(mod_w)
mod_s = np.array([])
for s in modern_ss:
    s = [x for x in s if x is not None]
    mod_s = np.concatenate((mod_s, s))
    
modern['words'] = mod_w
modern['books'] = mod_b
modern['sentences'] = mod_s
```

Now, with the final dataset assembled, with each era's dictionary containing the corresponding unique words, sentences, and books, we can serialize our 4 datasets so we can create the word embeddings in the next file.

```python
pickle.dump(classical, open('classical.pkl', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
pickle.dump(romantic, open('romantic.pkl', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
pickle.dump(victorian, open('victorian.pkl', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
pickle.dump(modern, open('modern.pkl', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
```
